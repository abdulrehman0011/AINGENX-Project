```python
pd.set_option('display.max_columns' , None)
df = pd.read_csv("/content/Diwali-Sale-Data.csv", encoding="ISO-8859-1")

df.head()
```

```python
# remove unnecessary column
# inspect and handle null value
```

```python
# remove unnamed1 column fron table
df  =  ps.remove(df , operation = 'columns' , columns = 'unnamed1')
df.head()
```

```python
# handle null values
#  to display all rows if rows show ...
pd.set_option('display.max_rows' , None)

df.isnull().sum()
```

```python
# replace null value "Null Values"

df =  ps.handle_nulls(df , columns = 'Status' , action = 'replace' , with_val = 'Null Values')
```

```python
df.head()
```

```python
# count category
df['Product_Category'].value_counts()
```

```python
# value_count for age
df['Age'].value_counts()
```

```python
# interquartile range (iqr)
df = ps.detect_outliers(df, method='iqr', threshold=1.5 , columns = 'Age')
```
